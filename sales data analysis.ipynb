# SALES DATA ANALYSIS

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sales_data = pd.read_csv("C:\\Users\\Navneet Raj\\Desktop\\sales data.csv")
```

```python
sales_data
```

```python
sales_data.info()
```

# Analyse sales trends overtime

```python
sales_data['Order Date'] = pd.to_datetime(sales_data['Order Date'])
```

```python
sales_data['Order Date']
```

# Calculate Total sales

```python
total_sales = sales_data['Sales'].sum()
total_sales
```

```python
total_cost = sales_data['Price Each'].sum()
total_cost
```

```python
total_profit = total_sales - total_cost
```

```python
total_profit
```

```python
profit_margin = (total_profit / total_sales)*100
```

```python
profit_margin
```

# Identify best selling products

```python
best_selling_products = sales_data.groupby('Product')['Quantity Ordered'].sum().sort_values(ascending=False)
```

```python
best_selling_products
```

# Visualizing the sales trends

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Month', y='Sales', data=sales_data)
plt.title('Sales Trend Over Time')
plt.show()
```

# Visualizing the Best-selling products

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=best_selling_products.index, y=best_selling_products.values)
plt.title('Best-Selling Products')
plt.xlabel('Product Name')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=50)
plt.show()
```
